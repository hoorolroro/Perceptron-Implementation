# Sigmoid 함수를 이용한 Perceptron 구현

```
import torch
import torch.nn as nn
import numpy as np

x_data = torch.FloatTensor([1.0, 2.0]) # x = (1, 2) 값 생성
print(x_data)

linear = nn.Linear(2, 1)  # y = (x_1w_1, x_2w_2) + b perceptron 생성
sigmoid = nn.Sigmoid()

z = sigmoid(linear(x_data))
print(z)
```

```
print(linear.weight)
print(linear.bias)
```

# Relu 함수를 이용한 Perceptron 구현

```
x_data= np.array([1.0, 2.0], dtype=np.float32)
x_data = torch.from_numpy(x_data)

linear = nn.Linear(2, 1)
relu = nn.ReLU()

z = relu(linear(x_data))

print(z)
print(linear.weight)
print(linear.bias)
```

질문 : z 값이 1에 가까울수록 좋은건가?
